```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as ppl
from sklearn.linear_model import LinearRegression as Model
from abc import ABC, abstractmethod
```

```python
file = './data2.csv'
```

```python
class DataProcessor:
    def __init__(self, r_train=0.6, r_cv=0.2):
        self.r_train = r_train
        self.r_cv = r_cv
    
    def load_data(self, file):
        rawData = pd.read_csv(file)
        (rows, cols) = rawData.shape
        self.X = rawData.iloc[:, : cols - 1]
        self.y = rawData.iloc[:, cols - 1]
        return self
    
    def plot_data(self):
        for column in self.X.columns:
            %matplotlib inline
            ppl.scatter(self.X[column], self.y)
            ppl.xlabel(column)
            ppl.ylabel(self.y.name)
            ppl.show()
        return self
            
    def split_data(self):
        n_total = self.X.shape[0]
        n_train = int(n_total * self.r_train)
        n_cv = int(n_total * (self.r_cv + self.r_train))
        idx = np.random.permutation(n_total)
        idx_train = idx[: n_train]
        idx_cv = idx[n_train : n_cv]
        idx_test = idx[n_cv :]
        self.X_train = self.X.iloc[idx_train, :]
        self.X_cv = self.X.iloc[idx_cv, :]
        self.X_test = self.X.iloc[idx_test, :]
        self.y_train = self.y[idx_train]
        self.y_cv = self.y[idx_cv]
        self.y_test = self.y[idx_test]
        return self
```

```python
dataProcessor = DataProcessor()
dataProcessor.load_data(file)
dataProcessor.plot_data()
dataProcessor.split_data()

print('X_train: {0} - X_cv: {1} - X_test: {2}'.format(dataProcessor.X_train.shape, dataProcessor.X_cv.shape, dataProcessor.X_test.shape))
```

```python
class ModelTrainerBase(ABC):
    def compute_errors(self, X, y):
        y_predict = self.predict(X)
        return y_predict - y
    
    def compute_cost(self, X, y):
        m = len(y)
        y_predict = self.predict(X)
        return np.sqrt(np.sum(y_predict - y) ** 2) / (2 * m)

    @abstractmethod
    def fit(self, X, y):
        pass
    
    @abstractmethod
    def predict(self, X):
        pass
    
    @abstractmethod
    def get_coefficients(self):
        pass
```

```python
class ScikitTrainer(ModelTrainerBase):
    def fit(self, X, y):
        self.model = Model(normalize=True, copy_X=True)
        self.model.fit(X, y)
        return self

    def predict(self, X):
        return self.model.predict(X)
    
    def get_coefficients(self):
        return np.append(self.model.intercept_, self.model.coef_)
```

```python
class ManualTrainer(ModelTrainerBase):
    def fit(self, X, y):
        (x_rows, x_cols) = X.shape
        X_processed = self._prepend_ones(self._normalize(X))
#         print(X[: 5])
        self._coefficients = self._initialize_coefficients(x_cols + 1)
        learning_rate = 0.1
        iterate_limit = 10000
        iterate = 0
        while True:
            new_coefs = self._run_gradient_descent(X_processed, y, self._coefficients, learning_rate)
            iterate += 1
            if iterate >= iterate_limit or np.allclose(self._coefficients, new_coefs, atol=1e-14, rtol=1e-14):
                break
            self._coefficients = new_coefs            
        
    def predict(self, X):
        X_processed = self._prepend_ones(self._zoom(X))
#         print(X_processed[: 5])
        return self._predict(X_processed)

    def get_coefficients(self):
        return self._coefficients
    
    def _predict(self, X, coefficients=None):
        if coefficients is None:
            coefficients = self._coefficients
        return np.dot(X, coefficients)
    
    def _initialize_coefficients(self, count):
        return np.random.random_sample((count, ))
        
    def _prepend_ones(self, matrix):
        (rows, cols) = matrix.shape
        ones = np.full(shape=(rows, 1), fill_value=1, dtype=float)
        return np.append(ones, matrix, axis=1)
    
    def _normalize(self, matrix):
        self._min = min = -1
        max = 1
        range = max - min
        self._matrix_min = matrix_min = np.min(matrix, axis=0)
        matrix_max = np.max(matrix, axis=0)
        matrix_range = matrix_max - matrix_min
        self._ratio = ratio = range / matrix_range
        return self._zoom(matrix, matrix_min, ratio, min)
    
    def _zoom(self, matrix, matrix_min=None, ratio=None, min=None):
        matrix_min = self._matrix_min if matrix_min is None else matrix_min
        ratio = self._ratio if ratio is None else ratio
        min = self._min if min is None else min
        return (matrix - matrix_min) * ratio + min
#         return matrix
        
    
    def _run_gradient_descent(self, X, y, coefficients, learning_rate):
        m = len(y)
        y_predict = self._predict(X, coefficients)
        error = y_predict - y
        deviation = np.dot(error, X) / m
#         print('X: {0} - y: {1} - coefficients: {2} - learning_rate: {3} - m: {4} - error: {5} - deviation: {6}'.format(X.shape, y.shape, coefficients.shape, learning_rate, m, error.shape, deviation.shape))
#         print('error: {0}'.format(error[: 5]))
#         print('coefs: {0}'.format(coefficients))
        return coefficients - learning_rate * deviation
```

```python
def testTrainer(Trainer):
    X_test = dataProcessor.X_test.iloc[: 5, :]
    y_test = dataProcessor.y_test[: 5]
    trainer = Trainer()
    trainer.fit(dataProcessor.X_train, dataProcessor.y_train)
    print()
    print('----------Test {0}----------'.format(trainer.__class__.__name__))
    print('coefficients: {0}'.format(trainer.get_coefficients()))
    print('y_predict: {0}'.format(trainer.predict(X_test)))
    print('y_test:\n{0}'.format(y_test))
    print('errors: {0}'.format(trainer.compute_errors(X_test, y_test)))
    print('cost: {0}'.format(trainer.compute_cost(X_test, y_test)))
    
testTrainer(ScikitTrainer)
testTrainer(ManualTrainer)
```
