```python
import numpy as np
import tensorflow as tf
from tensorflow.python.framework import ops
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Generating data

```python
SIZE = 400
SCALE = 15
CENTER_1 = [0.3, -23.1]
CENTER_2 = [36.3, 17.85]

x1_A = np.random.normal(loc=CENTER_1[0], scale=SCALE, size=(SIZE, 1))
x2_A = np.random.normal(loc=CENTER_1[1], scale=SCALE, size=(SIZE, 1))
y_A_0 = np.full(shape=[SIZE, 1], fill_value=0)
y_A_1 = np.full(shape=[SIZE, 1], fill_value=-1)

x1_B = np.random.normal(loc=CENTER_2[0], scale=SCALE, size=(SIZE, 1))
x2_B = np.random.normal(loc=CENTER_2[1], scale=SCALE, size=(SIZE, 1))
y_B = np.full(shape=[SIZE, 1], fill_value=1)

X_data = np.column_stack((np.row_stack((x1_A, x1_B)), np.row_stack((x2_A, x2_B))))
y_0_data = np.row_stack((y_A_0, y_B))
y_1_data = np.row_stack((y_A_1, y_B))

print(X_data.shape)
print(y_0_data.shape)
print(y_1_data.shape)
```

### Plotting data

```python
plt.scatter(x1_A, x2_A, marker='o', c='m')
plt.scatter(x1_B, x2_B, marker='x', c='c')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

## Using Linear Regression

### Resetting graph

```python
ops.reset_default_graph()
sess = tf.Session()
```

### Declaring placeholders and variables

```python
X = tf.placeholder(dtype=tf.float32, shape=[None, 2], name='X')
y = tf.placeholder(dtype=tf.float32, shape=[None, 1], name='y')

A = tf.Variable(tf.truncated_normal(mean=0, stddev=1, dtype=tf.float32, shape=[2, 1]), name='A')
b = tf.Variable(tf.truncated_normal(mean=0, stddev=1, dtype=tf.float32, shape=[1, 1]), name='b')

print(X)
print(y)
print(A)
print(b)
```

### Calculating output and loss

```python
output = tf.add(tf.matmul(X, A), b, name='output')
loss = tf.reduce_mean(tf.nn.sigmoid_cross_entropy_with_logits(labels=y, logits=output), name='loss')

print(output)
print(loss)
```

### Prediction and accuracy

```python
threshold = 0.5
prediction = tf.cast(tf.greater(tf.sigmoid(output), 0.5), dtype=tf.float32, name='prediction')
accuracy = tf.reduce_mean(tf.cast(tf.equal(prediction, y), dtype=tf.float32), name='accuracy')

print(prediction)
print(accuracy)
```

### Training step

```python
optimizer = tf.train.GradientDescentOptimizer(0.1)
train = optimizer.minimize(loss)
```

### Initializing variables

```python
init = tf.global_variables_initializer()
sess.run(init)
```

### Training

```python
ITERATIONS = 10000
LOG_STEP = ITERATIONS // 10
feed_dict = {X: X_data, y: y_0_data}

for i in range(ITERATIONS):
    sess.run(train, feed_dict=feed_dict)
    
    if i == 0 or (i + 1) % LOG_STEP == 0:
        loss_res = sess.run(loss, feed_dict=feed_dict)
        print('#{} - loss: {}'.format(i + 1, loss_res))
```

```python
[[a1_res], [a2_res]] = sess.run(A)
[[b_res]] = sess.run(b)
slope = -a2_res / a1_res
intercept = b_res / a1_res

print(a1_res)
print(a2_res)
print(b_res)
print(slope)
print(intercept)
```

### Accuracy

```python
sess.run(accuracy, feed_dict=feed_dict)
```

```python
x_line = np.linspace(-20, 50, 100)
y_line = x_line * slope + intercept

plt.scatter(x1_A, x2_A, marker='o', c='m')
plt.scatter(x1_B, x2_B, marker='x', c='c')
plt.plot(x_line, y_line)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

## Using Linear SVM

$$loss = \frac{1}{n}\displaystyle\sum^n_{i = 1}max\left(0, 1 -y_i(Ax_i - b)\right) + a \|A\|^2$$

### Declaring placeholders and variables

### Resetting graph

```python
ops.reset_default_graph()
sess = tf.Session()
```

### Declaring placeholders and variables

```python
X = tf.placeholder(dtype=tf.float32, shape=[None, 2], name='X')
y = tf.placeholder(dtype=tf.float32, shape=[None, 1], name='y')

A = tf.Variable(tf.truncated_normal(mean=0, stddev=1, dtype=tf.float32, shape=[2, 1]), name='A')
b = tf.Variable(tf.truncated_normal(mean=0, stddev=1, dtype=tf.float32, shape=[1, 1]), name='b')

print(X)
print(y)
print(A)
print(b)
```

### Calculating output and loss

```python
alpha = tf.constant(0.05, name='alpha')
output = tf.subtract(tf.matmul(X, A), b, name='output')
classification = tf.reduce_mean(tf.maximum(0., tf.subtract(1., tf.multiply(y, output))), name='classification')
loss = tf.add(classification, tf.multiply(alpha, tf.reduce_sum(tf.square(A))), name='loss')

print(alpha)
print(output)
print(classification)
print(loss)
```

### Prediction and accuracy

```python
prediction = tf.sign(output, name='prediction')
accuracy = tf.reduce_mean(tf.cast(tf.equal(prediction, y), dtype=tf.float32), name='accuracy')

print(prediction)
print(accuracy)
```

### Training step

```python
optimizer = tf.train.GradientDescentOptimizer(0.05)
train = optimizer.minimize(loss)
```

### Initializing variables

```python
init = tf.global_variables_initializer()
sess.run(init)
```

### Training

```python
ITERATIONS = 10000
LOG_STEP = ITERATIONS // 10
feed_dict = {X: X_data, y: y_1_data}

for i in range(ITERATIONS):
    sess.run(train, feed_dict=feed_dict)
    
    if i == 0 or (i + 1) % LOG_STEP == 0:
        loss_res = sess.run(loss, feed_dict=feed_dict)
        print('#{} - loss: {}'.format(i + 1, loss_res))
```

```python
[[a1_res], [a2_res]] = sess.run(A)
[[b_res]] = sess.run(b)
slope = -a2_res / a1_res
intercept = b_res / a1_res

print(a1_res)
print(a2_res)
print(b_res)
print(slope)
print(intercept)
```

### Accuracy

```python
sess.run(accuracy, feed_dict=feed_dict)
```

```python
x_line = np.linspace(-20, 50, 100)
y_line = x_line * slope + intercept

plt.scatter(x1_A, x2_A, marker='o', c='m')
plt.scatter(x1_B, x2_B, marker='x', c='c')
plt.plot(x_line, y_line)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
