```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
x = np.linspace(-10, 10, 1000)
```

```python
def plotFunction(fn, xRange=(-10 ,10)):
    x = np.linspace(xRange[0], xRange[1], 1000)
    y = fn(x)
    plt.figure(figsize=(8, 6))
    plt.plot(x, y)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()
```

```python
# plot softplus function
plotFunction(lambda x: np.log(1 + np.exp(x)))
```

```python
# plot sigmoide function
plotFunction(lambda x: 1 / (1 + np.exp(- x)), xRange=(-5, 5))
```
