```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
size = 100

x_standard = np.linspace(0, 1, 100)
y_standard = x_standard * 5 - 9
```

```python
x = np.random.rand(size)
```

```python
y = 5 * x - 9
```

```python
plt.scatter(x, y, marker='x')
plt.xlabel('X')
plt.ylabel('Y')
plt.plot(x, y, 'r-')
```

```python
yy = np.vectorize(lambda y: y + np.random.normal(loc=0, scale=0.2))(y)
```

```python
plt.scatter(x, yy, marker='x')
plt.xlabel('X')
plt.ylabel('YY')
plt.plot(x, y, 'r-')
```

```python
noise = np.random.normal(loc=0, scale=0.25, size=size)
yyy = y + noise
```

```python
plt.scatter(x, yyy, marker='x')
plt.xlabel('X')
plt.ylabel('YYY')
plt.plot(x, y, 'r-')
```
