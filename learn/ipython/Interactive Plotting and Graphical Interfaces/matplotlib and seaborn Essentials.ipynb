## Common plots with ``matplotlib``

A line plot represents a mathematical curve or a digital signal as a continuous succession of line segments.

```python
y = np.random.randn(1000)
plt.plot(y)
```

Plot the graph of a mathematical function:

```python
x = np.linspace(-10., 10., 1000)
y = np.sin(3 * x) * np.exp(-0.1 * x ** 2)
plt.plot(x, y)
```

All aspects of the plot can be customized

```python
x = np.linspace(-5, 5, 100)
y = np.sin(3 * x) * np.exp(-0.1 * x ** 2)
plt.plot(x, y, '-.o', lw=2, color='#FDBB84', mfc='#2B8CBE', ms=8)
```

* The third argument is a format string specifying the aspect of the plot in a compact form:
    * The `--` characters indicate a dashed line style.
    * The `^` character indicates a upper triangle marker.
* The `lw` argument indicates the line width.
* The color can be specified in many ways, including as a hexadecimal RGB value.
* The marker face color is indicated by `mfc`.
* The marker size is indicated by `ms`.

You will find more details about the plot customization options at http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.plot.

## Customizing ``matplotlib`` figures

We can create multiple plots in the same figure.

```python
# Left panel
plt.subplot(1, 2, 1)
x = np.linspace(-10.0, 10.0, 1000)
plt.plot(x, np.sin(x), '--m', label='sin')
plt.plot(x, np.cos(x), ':c', label='cos')
plt.xticks([-10, 0, 10])
plt.yticks([-1, 0, 1])
plt.ylim(-2, 2)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('SIN & COS')
plt.legend()

# Right panel
plt.subplot(1, 2, 2, polar=True)
x = np.linspace(0, 2 * np.pi, 1000)
plt.plot(x, 1 + 2 * np.cos(6 * x))
plt.yticks([])
plt.xlim(-0.1, 3.1)
plt.ylim(-0.1, 3.1)
plt.xticks(np.linspace(0, 5 * np.pi / 3, 6))
plt.title('POLAR PLOT')
plt.grid(color='g', linewidth=1, linestyle=':')
```

You can also customize the defaults of matplotlib. Refer to the following link for more information: http://matplotlib.org/users/customizing.html.

You will find hundreds of examples in the official matplotlib gallery at http://matplotlib.org/gallery.html.

## Interacting with ``matplotlib`` figures in the Notebook

You can update matplotlib figures interactively in the Notebook using widgets.

```python
import matplotlib
matplotlib.use('nbagg')

import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact
x = np.linspace(-5.0, 5.0, 1000)

@interact(a=(1, 10))
def plot_sin(a):
    plt.plot(x, np.sin(a * x))
    plt.ylim(-1, 1)
    plt.show()
```
